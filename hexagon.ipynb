```python
def hexagon(coord0):
    import math
    a = 1.418
    list_of_coords = []
    
    for i in range(6):
        angle_rad = (2 * i + 1) * math.pi / 6
        r = [a * math.cos(angle_rad), a * math.sin(angle_rad), 0]
        coord = [round(x0 + x, 3) for x0, x in zip(coord0, r)]
        list_of_coords.append(coord)

    return list_of_coords
```

```python
def generate_pristine_graphene_hexagon(n):
    import math
    coord0 = [0, 0, 0]
    list_of_coords = []
    a0 = 1.418
    a = 2 * a0 * math.cos(math.pi / 6)
    for i in range(n):
        for j in range(6):
            angle_rad = j * 2*math.pi/6
            r = [i * a * math.cos(angle_rad), i * a * math.sin(angle_rad), 0]
            coord = [x0 + x for x0, x in zip(coord0, r)]
            list_of_coords.extend(hexagon(coord))
            angle_rad_k = angle_rad + 4*math.pi/3
            for k in range(1, i):
                r_k = [k * a * math.cos(angle_rad_k), k * a * math.sin(angle_rad_k), 0]
                coord_k = [x0 + x for x0, x in zip(coord, r_k)]
                list_of_coords.extend(hexagon(coord_k))

    list_of_coords = [list(x) for x in {tuple(coords) for coords in list_of_coords}]
    
    return list_of_coords
```
