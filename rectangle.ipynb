```python
def hexagon(coord0):
    import math
    a0 = 1.418
    list_of_coords = []
    
    for i in range(6):
        angle_rad = (2 * i + 1) * math.pi / 6
        r = [a0 * math.cos(angle_rad), a0 * math.sin(angle_rad), 0]
        coord = [round(x0 + x, 3) for x0, x in zip(coord0, r)]
        list_of_coords.append(coord)

    return list_of_coords
```

```python
def generate_pristine_graphene_rectangle(n, m, k):
    import math
    coord0 = [0, 0, 0]
    list_of_coords = []
    a0 = 1.418
    a = 2 * a0 * math.cos(math.pi / 6)

    for i in range(0, m, 2):
        for j in range(n):
            r = [i * 3/2 * a0, j * a, 0]
            r = [j * a, i * 3/2 * a0, 0]
            coord = [x0 + x for x0, x in zip(coord0, r)]
            list_of_coords.extend(hexagon(coord))

    for i in range(1, m, 2):
        for j in range(n + k):
            sign = (k > 0) - (k <= 0)
            r = [i * 3/2 * a0, (j - sign * 1/2) * a, 0]
            r = [(j - sign * 1/2) * a, i * 3/2 * a0, 0]
            coord = [x0 + x for x0, x in zip(coord0, r)]
            list_of_coords.extend(hexagon(coord))


    list_of_coords = [list(x) for x in {tuple(coords) for coords in list_of_coords}]
    
    return list_of_coords
```

```python
print(len(generate_pristine_graphene_rectangle(2, 2, -1)))
```
